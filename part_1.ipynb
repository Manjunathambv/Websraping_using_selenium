```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.common.exceptions import TimeoutException
import time
import ast
```

```python
service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service = service)
```

```python
driver.get("https://www.instagram.com/")
```

```python
"""
Problem 1 : Login to your Instagram
Login to your Instagram Handle
Submit with sample username and password
"""
#update your username and password here
username = 'SAMPLE USERNAME'
password = 'SAMPLE PASSWORD'

#initializing  wait object
wait = WebDriverWait(driver, 10)
def LogIn(username, password):
    try :
        #locating username textbox and sending username
        user_name = wait.until(EC.presence_of_element_located((By.NAME,'username')))
        user_name.send_keys(username)
        #locating password box and sending password
        pwd = driver.find_element(By.NAME, 'password')
        pwd.send_keys(password)
        #locating login button 
        button = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="loginForm"]/div[1]/div[3]/button/div')))
        button.submit()
        #Save Your Login Info? : Not Now
        pop = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_ac8f')))
        pop.click()
        
    except TimeoutException :
        print ("Something went wrong! Try Again")

```

```python
"""
Problem 2 : Type for “food” in search bar
Type for “food” in search bar and print all the names of the 
Instagram Handles that are displayed in list after typing “food”
Note : Make sure to avoid printing hashtags
"""
def search(s):
    list_food = []
    #locating serch bar and sending text
    search_bar = driver.find_element(By.XPATH, "//input[@placeholder='Search']")
    time.sleep(3)
    # send an argument food (based on the objective)
    search_bar.send_keys(s)
    time.sleep(3)
    
    #extract the profile related to the food
    list_ = driver.find_elements(By.XPATH, '//a[contains(@href, "food")]/div/div/div/div/div/div/div//*')
    for li in list_:
        t = (li.text)
        if t != "" : 
            list_food.append(t)

    return list_food

name_list = search("food")
for name in name_list:
    print(name)
```

```python
"""
Problem 3 : Searching and Opening a profile
Searching and Opening a profile using
Open profile of “So Delhi”
"""
def search_open_profile(s):
    try:
        #locatong search box bar and sending text 
        wait = WebDriverWait(driver, 10)
        search_box = wait.until(EC.presence_of_element_located((By.XPATH, "//input[@placeholder='Search']")))
        time.sleep(3)

        search_box.send_keys(s)
        #locating serched result
        time.sleep(3)
        res = wait.until(EC.presence_of_element_located((By.XPATH, "//div/div[contains(@class, 'x9f619')]/a[contains(@href, '/')]/div")))
        res.click()           

        time.sleep(5)
        #driver.back()

    except TimeoutException :
        print('No Search Found!')
        
```

```python
"""
Problem 4 : Follow/Unfollow given handle
Follow/Unfollow given handle -
1.Open the Instagram Handle of “So Delhi”
2.Start following it. Print a message if you are already following
3.After following, unfollow the instagram handle. Print a message if you have already unfollowed.
"""

def follow():
    try :
        #locating follow button
        btn = wait.until(EC.presence_of_element_located((By.XPATH, "//button[contains (@class, '_acan')]")))
        #checking for text
        if btn.text == 'Follow' : 
            btn.click()
            time.sleep(3)
        
        else : 
            print('Already Following')

    except TimeoutException :
        print("Something Went Wrong!")
        
def unfollow():
    try :
        #locating follow button
        btn = wait.until(EC.presence_of_element_located((By.XPATH, "//button[contains (@class, '_acan')]")))
        #checking for text
        if btn.text !='Follow' : 
            btn.click()
            time.sleep(2)
            #locating popup window (when you click on follow button)
            pop_up = wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains (@role, 'button')][5]")))
            pop_up.click()
            time.sleep(3)
            
        else : 
            print('Already Unfollowed')

    except TimeoutException :
        print("Something Went Wrong!")
        
search_open_profile('So Delhi')
follow()
unfollow()
```

```python
"""
Problem 5 : Like/Unlike posts
Like/Unlike posts
1.Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
2.Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.
"""

def Like_Post():
    try :
        
        #scrolling for locating post
        driver.execute_script('window.scrollTo(0, 6000);')
        time.sleep(3)
        driver.execute_script('window.scrollTo(0, -6000);')
        time.sleep(3)
        
        click_button = driver.find_elements(By.XPATH, "//div[contains (@class, '_aabd')]/a")
        
        for i in range(30):
            click_button[i].click()
            time.sleep(2)
            #locating like/unke button
            like = wait.until(EC.presence_of_element_located((By.XPATH, "//span[contains (@class, '_aamw')]//div[contains (@role, 'button')]//span")))
#             st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            time.sleep(3)
            st = ast.literal_eval(like.get_attribute("innerHTML").split()[1].split("=")[1])
            if st == 'Like' :
                like.click()
                print('You LIKED :', i+1)
                time.sleep(2)
            else :
                print('You have already LIKED Post Number :', i+1)
                time.sleep(2)
            #locating cross button for closing post
            close_button = driver.find_element(By.XPATH, "//div[contains (@class, 'x160vmok')]//div[contains (@role, 'button')]")
            close_button.click()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
        
def Unlike_Post():
    try :
        
        #scrolling for locating post
        driver.execute_script('window.scrollTo(0, 6000);')
        time.sleep(3)
        driver.execute_script('window.scrollTo(0, -6000);')
        time.sleep(3)
        
        click_button = driver.find_elements(By.XPATH, "//div[contains (@class, '_aabd')]/a")
        
        for i in range(30):
            click_button[i].click()
            time.sleep(2)
            #locating like/unke button
            like = wait.until(EC.presence_of_element_located((By.XPATH, "//span[contains (@class, '_aamw')]//div[contains (@role, 'button')]//span")))
#             st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            time.sleep(3)
            st = ast.literal_eval(like.get_attribute("innerHTML").split()[1].split("=")[1])
            if st == 'Unlike' :
                like.click()
                print('You UNLIKED :', i+1)
                time.sleep(2)
            else :
                print('You have already UNLIKED Post Number :', i+1)
                time.sleep(2)
            #locating cross button for closing post
            close_button = driver.find_element(By.XPATH, "//div[contains (@class, 'x160vmok')]//div[contains (@role, 'button')]")
            close_button.click()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
"""
Problem 6 : Extract list of followers
Extract list of followers
1.Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
2.Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.
"""
def Extract_Followers():
    try :
        followers_btn = driver.find_element(By.XPATH, "//ul[contains (@class, 'x78zum5')]/li/a")
        time.sleep(3)
        followers_btn.click()
        
        time.sleep(3)
        frame = driver.find_element(By.XPATH, "//div[contains (@class, '_aano')]")
        for i in range(50):
            time.sleep(1)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",frame)
        
        usernames = driver.find_elements(By.XPATH, "//div[contains (@class, 'x1rg5ohu')]//a")
        
        names = []
        for username in usernames[:500]:
            names.append(username.text)
            
        return names
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
"""
Problem 7 : Check the story of ‘coding.ninjas’
Check the story of ‘coding.ninjas’. Consider the following Scenarios 
and print error messages accordingly -
1.If You have already seen the story.
2.Or The user has no story.
3.Or View the story if not yet seen.
"""
def stroy_check():
    profile_link = driver.find_element(By.XPATH, "//img[contains (@alt, 'coding')]")
    profile_link.click()

    story_btn = driver.find_element(By.XPATH,'//div[contains (@class,"_aarf")]')
    #check is story is present using aria-disabled attribute

    if story_btn.get_attribute("aria-disabled") == "true":
        print("Coding Ninjas has no story")

    else:
        #if story is present check if you have viewed the story by checking height attribute 
        story_check = driver.find_element(By.XPATH,'//div[contains(@class,"_aarf")]/canvas')
        if story_check.get_attribute("height") == '166':
            print("I have already seen the story")


        #if not viewed then view the story
        else:
            print("Viewing the story")
            viewstory = driver.find_element(By.XPATH,'//div[starts-with(@class,"_aarf")]/span')
            viewstory.click()
            time.sleep(4)
            while True:
                try:
                    next_story = driver.find_element(By.XPATH,'//button[@aria-label="Next"]')
                    next_story = driver.find_element(By.XPATH,'//button[@aria-label="Next"]')
                    next_story.click()
                    time.sleep(4)
                except:
                    break
            
```
